##1. Cosine Similarity Using TF-IDF
Cosine similarity measures the cosine of the angle between two vectors, which can be used to compare the similarity between the resume and job description text.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([resume_text, job_description])

# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print("Cosine Similarity:", cosine_sim[0][0])
```

##2. BERT Embeddings with Cosine Similarity
BERT (Bidirectional Encoder Representations from Transformers) is a powerful model for generating context-aware embeddings for text. We can use it to obtain embeddings for both the resume and job description and then calculate their cosine similarity.



```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Generate BERT embeddings
resume_embedding = model.encode(resume_text, convert_to_tensor=True)
job_embedding = model.encode(job_description, convert_to_tensor=True)

# Calculate cosine similarity
cosine_sim = util.pytorch_cos_sim(resume_embedding, job_embedding)
print("Cosine Similarity:", cosine_sim.item())
```

##3. Jaccard Similarity
Jaccard similarity measures the similarity between two sets by comparing the intersection and union of the sets. This can be applied to tokenized text.

```python
def jaccard_similarity(text1, text2):
    # Tokenize the texts
    tokens1 = set(text1.lower().split())
    tokens2 = set(text2.lower().split())

    # Calculate intersection and union
    intersection = tokens1.intersection(tokens2)
    union = tokens1.union(tokens2)

    # Calculate Jaccard similarity
    similarity = len(intersection) / len(union)
    return similarity

# Calculate Jaccard similarity
similarity = jaccard_similarity(resume_text, job_description)
print("Jaccard Similarity:", similarity)
```

##4. Latent Semantic Analysis (LSA)
LSA is a technique in NLP that helps uncover the hidden relationships between words in documents by reducing the dimensionality of the term-document matrix.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

# Sample data
documents = [
   resume_text, job_description
]

# Create TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Apply LSA
lsa = TruncatedSVD(n_components=2)
lsa_matrix = lsa.fit_transform(tfidf_matrix)

# Calculate cosine similarity
cosine_sim = cosine_similarity(lsa_matrix[0:1], lsa_matrix[1:2])
print("Cosine Similarity after LSA:", cosine_sim[0][0])
```

##5. Keyword Matching with Weights
A simpler, rule-based approach involves assigning weights to keywords in both the resume and job description, then summing these weights to find a match score.

```python
from collections import Counter

def keyword_matching(resume_text, job_description, weights=None):
    if weights is None:
        weights = {}

    # Tokenize and count word frequencies
    resume_tokens = Counter(resume_text.lower().split())
    job_tokens = Counter(job_description.lower().split())

    # Calculate the match score based on weights
    score = 0
    for word in job_tokens:
        if word in resume_tokens:
            score += resume_tokens[word] * weights.get(word, 1)

    return score

job_description = "Looking for a software engineer with experience in full stack development and cloud technologies."
weights = {"software": 2, "engineer": 2, "development": 1.5, "cloud": 1.5}

# Calculate match score
match_score = keyword_matching(resume_text, job_description, weights)
print("Keyword Matching Score:", match_score)
```
